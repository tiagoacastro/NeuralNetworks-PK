### Create an image as a 2D array and display it with pyplot

#### Check board

```python
import numpy as np
import matplotlib.pyplot as plt
check = np.zeros((9, 9)) 
check[::2, 1::2] = 1 #here the pixels are from [0,1], i.e., 0 is black, 1 is white
check[1::2, ::2] = 1
print('shape',check.shape)
print('type', check.dtype) #note what is the default type of elements
plt.imshow(check, cmap='gray', interpolation='nearest') #note that pyplot display a 9x9 image as a "bigger image"
plt.show()
```

#### Simple gradient image

```python
import numpy as np
import matplotlib.pyplot as plt

im = np.zeros((300, 400), dtype='uint8') #here the pixel values are from [0,255], 0 is black, 255 is white
#make im a gradient image
rows, columns = im.shape
for i in range(columns):
    im[:,i] = i/columns*255
print('shape',im.shape)
print('type', im.dtype)
plt.imshow(im, cmap='gray', interpolation='nearest') 
plt.show()
```

### Task 5

#### Create an image with three channels (red, green, blue)

```python
import numpy as np
import matplotlib.pyplot as plt

im = np.zeros((300, 400, 3), dtype='uint8') #here the pixel values are from [0,255], 0 is black, 255 is white
rows, columns, channels = im.shape

print('shape',im.shape)
print('type', im.dtype)

im[:,:,0] = 255 #set the red channel to maximum
plt.imshow(im, cmap='gray', interpolation='nearest') 

plt.figure() # a new figure
im[:,:,0] = 0
im[:,:,1] = 255 #set the green channel to maximum
plt.imshow(im, cmap='gray', interpolation='nearest') 

plt.figure()
im[:,:,1] = 0
im[:,:,2] = 255 #set the blue channel to maximum
plt.imshow(im, cmap='gray', interpolation='nearest') 

plt.show()
```

### Task 8
How to create the following image?


```python
import numpy as np
import matplotlib.pyplot as plt

im = np.zeros((300, 400, 3), dtype='uint8') #here the pixel values are from [0,255], 0 is black, 255 is white
rows, columns, channels = im.shape

print('shape',im.shape)
print('type', im.dtype)

#magenta
im[0:rows//2,0:columns//2,0] = 255 #set the red channel to maximum
im[0:rows//2,0:columns//2,2] = 255 #set the blue channel to maximum

#blue
im[0:rows//2,columns//2:columns,2] = 255 #set the blue channel to maximum

#green
im[rows//2:rows,0:columns//2,1] = 255 #set the green channel to maximum

#cyan
im[rows//2:rows,columns//2:columns,1] = 255 #set the green channel to maximum
im[rows//2:rows,columns//2:columns,2] = 255 #set the blue channel to maximum

plt.imshow(im, cmap='gray', interpolation='nearest') 

plt.show()
```

### Task 10

How to display the following image? The convention is that if the pixel values are represented as floats, the minimum pixel value (black) is 0, and the maximum (white) is 1.


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img=mpimg.imread('lena_grayscale.png')
rows, columns, channels = img.shape
img[0:rows//2,0:columns//2,0:3] = 0
img[rows//2:rows,columns//2:columns,0:3] = 1

print('lena shape',img.shape) #note, how many channels are there?
print('lena dtype',img.dtype)
plt.imshow(img)
plt.show()
```

### Load a color image
